## plot for alpha values

```python
import matplotlib.pyplot as plt

# --- Hard‑coded α‑estimates ---
alpha_all     = {"PULSCAR": 0.0171, "PULSNAR": 0.0257}
alpha_rna     = {"PULSCAR": 0.0081, "PULSNAR": 0.0126}

# --- Create a 1×2 subplot figure ---
fig, axes = plt.subplots(1, 2, figsize=(8, 4), sharey=True)

# Left panel: all‑ligand balanced set
axes[0].bar(alpha_all.keys(), alpha_all.values(), color=["skyblue", "salmon"])
axes[0].set_title("Balanced (All Ligands)")
axes[0].set_ylabel("Estimated α")
axes[0].set_ylim(0, 0.03)  # add a bit of headroom

# Right panel: RNA‑specific balanced set
axes[1].bar(alpha_rna.keys(), alpha_rna.values(), color=["skyblue", "salmon"])
axes[1].set_title("Balanced (RNA‑specific)")
axes[1].set_ylim(0, 0.03)  # keep the same scale for easy comparison

# --- Tidy up and save/show ---
fig.suptitle("α Estimates from PULSCAR vs PULSNAR", fontsize=12, y=1.05)
fig.tight_layout()
plt.show()          # or fig.savefig("alpha_estimates.png", dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ─── Hard‑coded α values ──────────────────────────────────────────────────────
data = {
    "Manipulated (All Ligands)":   {"True α": 0.12230, "PULSCAR": 0.1457, "PULSNAR": 0.1490},
    "Manipulated (RNA‑specific)": {"True α": 0.14667, "PULSCAR": 0.1474, "PULSNAR": 0.1684},
}

# ─── Plot settings ────────────────────────────────────────────────────────────
labels = list(next(iter(data.values())).keys())         # ["True α", "PULSCAR", "PULSNAR"]
datasets = list(data.keys())                            # two dataset names
x = np.arange(len(labels))                              # bar locations
width = 0.35                                            # bar width

# ─── Create a 1×2 subplot figure ──────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(9, 4), sharey=True)

for ax, (title, values) in zip(axes, data.items()):
    bars = ax.bar(x, values.values(), width, color=["lightgray", "skyblue", "salmon"])
    
    ax.set_title(title, fontsize=10)
    ax.set_xticks(x)
    ax.set_xticklabels(labels, rotation=45, ha="right")
    ax.set_ylabel("Estimated α")
    ax.set_ylim(0, 0.20)
    
    # Optional: annotate bars with exact values
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{height:.3f}", xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords="offset points",
                    ha='center', va='bottom', fontsize=8)

fig.suptitle("α Estimates on Manipulated Datasets (Known Proportions)", y=1.02, fontsize=12)
fig.tight_layout()
plt.show()           # or: fig.savefig("alpha_estimates_manipulated.png", dpi=300)
```
